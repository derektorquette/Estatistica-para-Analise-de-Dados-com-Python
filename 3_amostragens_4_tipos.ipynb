## **AMOSTRAGEM ESTRATIFICADA** - Divisão da população em grupos e seleção aleatória de uma amostra de cada grupo. (Ex: divisão por reigião, classe social, religião, etc.)

```
enem_campinas['RACA'].value_counts() # contagem
```

```
enem_campinas['TP_LINGUA'].value_counts() # contagem
```

**Para quantidade pequena de grupos** - amostragem simples e aleatória

- lembrando que esse conceito de 'raça' é retirado do IBGE, essa classificação gera controvérsias

```
branca = enem_campinas.loc[enem_campinas.RACA == 'branca'].sample(frac=0.15) # fração de 15% na coluna 'Raça'
branca.shape # quantidade
```

```
parda = enem_campinas.loc[enem_campinas.RACA == 'parda'].sample(frac=0.15)
parda.shape
```

```
preta = enem_campinas.loc[enem_campinas.RACA == 'preta'].sample(frac=0.15)
preta.shape
```

```
amarela = enem_campinas.loc[enem_campinas.RACA == 'amarela'].sample(frac=0.15)
amarela.shape
```

```
nd = enem_campinas.loc[enem_campinas.RACA == 'nao_declarado'].sample(frac=0.15)
nd.shape
```

```
indigena = enem_campinas.loc[enem_campinas.RACA == 'indígena'].sample(frac=1)
indigena.shape
```

```
indigena
```

```
amos_estra = pd.concat([branca, parda, preta, amarela, nd, indigena]) # concatenação em uma única tabela em outra variável
```

```
amos_estra
```

**Grandes quantidades de grupos** - método de machine/deep learning
